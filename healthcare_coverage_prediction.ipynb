```python
#making a copy of original dataset
df_tmp = df.copy()
```

```python
#dropping columns with least relation with target variable and creating bias
df_tmp.drop(columns=["ID","PREFIX","SUFFIX","ZIP"],inplace=True,axis=1)
```

```python
df_tmp
```

```python
Q1 = df_tmp.HEALTHCARE_COVERAGE.quantile(0.25)
Q3 = df_tmp.HEALTHCARE_COVERAGE.quantile(0.75)
Q1,Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit,upper_limit
```

```python
df_tmp[(df_tmp.HEALTHCARE_COVERAGE<lower_limit)|(df_tmp.HEALTHCARE_COVERAGE>upper_limit)]
```

```python
df_tmp =df_tmp[(df_tmp.HEALTHCARE_COVERAGE>lower_limit)&(df_tmp.HEALTHCARE_COVERAGE<upper_limit)]
```

```python
df_tmp
```

```python
# turn categorical variables into numbers and fill missing
for label,content in df_tmp.items():
    if not pd.api.types.is_numeric_dtype(content):
        #Turn categories into numbers and add +1
        df_tmp[label] = pd.Categorical(content).codes+1 #taking care of NaN values
```
