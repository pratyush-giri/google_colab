```python
image_paths= []
for i in range(1,11):
    if i !=10:
        my_file = open(f"./lfw_funneled/pairs_0{i}.txt", "r")
    else:
        my_file = open(f"./lfw_funneled/pairs_{i}.txt", "r")

    # reading the file
    data = my_file.read()

    data_into_list = data.split("\n")
    image_paths.extend(data_into_list)
    my_file.close()
```

```python
my_file = open(f"./lfw_funneled/pairs.txt", "r")
data = my_file.read()
data = data.split("\n")
data.pop(0)
len(data)
```

```python
pattern = "^([a-zA-Z]{0,}\_[a-zA-Z]{0,})"
import re
names = []
for name in data:
    names.extend(re.findall(pattern,name))
names.sort()
len(names)
```
